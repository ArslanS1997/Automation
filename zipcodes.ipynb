#### Looping Through all zipcodes
This loops through all the zipcodes in Zips.csv and outputs answer in Result.csv

```python
data = pd.read_csv('Result.csv')
#print(data['0'].count())
Avail = [x for x in data['TV_Avail']]

LIMIT =1000
length =data['TV_Avail'].count()-1
#print(length)
Current =0
ErrorCount =0
Curr =""
for zipcode in df['Zips'][length+1:]:
    Curr =""
    #dummy =0
    try:
        Curr = CheckAvail(zipcode)
       # print(Curr)
        Avail.append(Curr)
    except:
        Avail.append("ERROR")
        ErrorCount+=1
        
        #Curr ="ERROR"
        
    
    Current+=1
    if(Current ==LIMIT or ErrorCount>0):
        print('DONE ',Current,"Errors",ErrorCount," ",Curr)
        break
    if(Current%10==0):
        #if(len(str(zipcode))<5):
            #print('DONE ',Current,"Errors",ErrorCount," ",Curr,"Zip ","0"+str(zipcode))
            
        #print('DONE ',Current,"Errors",ErrorCount," ",Curr,"Zip ",zipcode)
#         yes_count=0
#         no_count=0
        ErrorCount=0
#     print('DONE ',Current)
    #print(zipcode)
    print('DONE ',Current,"Errors",ErrorCount," ",Curr,"Zip ",zipcode)
    
        
    
Avail = pd.Series(Avail)
#display(Avail)
tdf = pd.DataFrame({'TV_Avail' : Avail})
tdf.to_csv("Result.csv",index=False)
display(tdf)
```
